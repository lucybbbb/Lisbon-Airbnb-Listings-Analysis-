```python
# Lisbon Airbnb Data Cleaning Project
# Goal: Clean and prepare Airbnb listings for analysis in Power BI
# Load data
import pandas as pd
df = pd.read_excel(r"E:\projects data\lisbon-airbnb-analytics\data\lisbon_airbnb_listings.xlsx")

# Understand your data
df.head()

df.info()

df.columns

df.shape

# Cleaning
# Check missing values
df.isnull().sum()

#Remove duplicates
df = df.drop_duplicates()

#Clean column names
df.columns = df.columns.str.lower().str.replace(" ", "_")

#Quick validation
df.describe()

#Save cleaned data
df.to_excel('cleaned_airbnb.xlsx', index =False)
```
